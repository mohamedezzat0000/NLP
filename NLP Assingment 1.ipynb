```python
import requests
from bs4 import BeautifulSoup

def read_text_from_url(url):
    
    # Fetch HTML content from the URL
    response = requests.get(url)

    # Parse HTML content using BeautifulSoup
    content = BeautifulSoup(response.content, 'html.parser')

    # Extract text from HTML
    text = content.get_text()

    return text
```

```python
url = 'https://www.britannica.com/topic/BMW'
html_text = read_text_from_url(url)
print(html_text)
```

```python
import re 
html_text= re.sub(r'[^a-zA-Z0-9\s]', '', html_text)
html_text = re.sub('\s+',' ',html_text)
html_text=html_text.lower()
html_text
```

## Tokeniztion by splitting text

```python
html_text1= html_text.split()
html_text1
```

                                         
                                         
   ## Tokenization from NLTK data


```python
import nltk
from nltk.tokenize import word_tokenize

# Download NLTK data
nltk.download('punkt')
```

```python
tokens = nltk.word_tokenize(html_text)
tokens
```

## Stemming

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('english')

# Initialize Porter stemmer
stemmer = PorterStemmer
```

## Stemmed words without stop-words

```python
stemmed_tokens=[stemmer().stem(token)for token in tokens if token not in stop_words]
stemmed_tokens
```

## unique_words_stemming

```python
unique_words_stemming = set(stemmed_tokens)
```

```python
unique_words_stemming
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
# initialize WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

## lemmatized words

```python
lemmatized_words = [lemmatizer.lemmatize(token) for token in tokens]
lemmatized_words
```

## unique_words_lemmatized ## 

```python
unique_words_lemmatized = set(lemmatized_words)
```

```python
unique_words_lemmatized
```
